```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag
from nltk import wordnet
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

```

```python
# make sentences
text = "Natural Language Processing (NLP) is a fascinating field. It combines linguistics, computer science, and artificial intelligence to enable computers to understand human language."
```

Part 1: tokenization

```python
# perform sent tokenization
sentences = sent_tokenize(text)
print("Sentences:")
for sentence in sentences:
    print(sentence)
# perform word tokenization
words = word_tokenize(text)
print("\nWords:")
print(words)
# count total number of sentences
num_sentences = len(sentences)
print("\nTotal number of sentences:", num_sentences)
# count total number of words
num_words = len(words)
print("Total number of words:", num_words)
```

Part 2: stop words

```python
# load english stopwords from nltk
stop_words = set(stopwords.words('english'))
# covernt the list of words to lowercase
lower_words = [word.lower() for word in words]
# remove stopwords from the list of words
filtered_words = [word for word in lower_words if word not in stop_words]
print("\nFiltered Words (without stopwords):")
print(filtered_words)
# compare the lenght of the original list of words with the filtered list
print("\nOriginal number of words:", len(words))
print("Filtered number of words:", len(filtered_words))

```

Part3: stemming

```python
# import the stemmer from nltk
from nltk.stem import PorterStemmer
# apply stemmer to each word after stop words
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
print("\nStemmed Words:")
print(stemmed_words)
# store stop words in new list 
stop_words_list = list(stop_words)
# print original word and its stemmed
print("\nOriginal Word - Stemmed Word")
for original, stemmed in zip(filtered_words, stemmed_words):
    print(f"{original} - {stemmed}")
```

part 4: lemmatization

```python
# import the wordnet lemmatizer from nltk
from nltk.stem import WordNetLemmatizer
# lemmatize each word after stop words
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print("\nLemmatized Words:")
print(lemmatized_words)
# store lemmatized words in new list
lemmatized_words_list = list(lemmatized_words)
# print original word and its lemmatized form from side by side
print("\nOriginal Word - Lemmatized Word")
for original, lemmatized in zip(filtered_words, lemmatized_words):
    print(f"{original} - {lemmatized}")
# compare the stemming and lemmatization results
print("\nOriginal Word - Stemmed Word - Lemmatized Word")
for original, stemmed, lemmatized in zip(filtered_words, stemmed_words, lemmatized_words):
    print(f"{original} - {stemmed} - {lemmatized}")

```
